## Drag Coefficient for steady cylinder

```python
import pylab
import numpy
import vtk
```

```python
%pylab inline
```

```python
arr=vtk.vtkStructuredGridReader()
arr.SetFileName("Figures/Benchmarks/vtk_steady_state.vtk")
arr.Update()
```

```python
arr_imm=vtk.vtkPolyDataReader()
arr_imm.SetFileName("Figures/Benchmarks/vtk_steady_state_imm.vtk")
arr_imm.Update()
```

```python
grid=arr.GetOutput()
points=grid.GetPoints()
data=grid.GetPointData()
data.Update()
```

```python
grid_imm=arr_imm.GetOutput()
points_imm=grid_imm.GetPoints()
data_imm=grid_imm.GetPointData()
data_imm.Update()
```

```python
print data.GetArrayName(0),data.GetArrayName(1)
```

```python
dims=grid.GetDimensions()
density=data.GetArray("density")
velocity=data.GetArray("velocity")
```

```python
numpoints_imm=grid_imm.GetNumberOfPoints()
```

```python
print numpoints_imm
```

```python
dims
```

```python
dens=numpy.zeros((dims[1],dims[0]))
velx=numpy.zeros((dims[1],dims[0]))
vely=numpy.zeros((dims[1],dims[0]))
```

```python
for i in range(0,dims[0]*dims[1]):
    y=i/dims[0]
    x=i%dims[0]
    dens[y,x]=density.GetTuple1(i)
    vx,vy,vz=velocity.GetTuple3(i)
    velx[y,x]=vx
    vely[y,x]=vy
```

```python
x_imm=numpy.zeros(numpoints_imm)
y_imm=numpy.zeros(numpoints_imm)
```

```python
print x_imm.size,y_imm.size
```

```python
for i in range(0,numpoints_imm):
    xp,yp,zp=points_imm.GetPoint(i)
    x_imm[i]=xp
    y_imm[i]=yp
```

```python
print dens,velx
```
